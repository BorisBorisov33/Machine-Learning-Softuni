```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

from sklearn.cluster import KMeans,AgglomerativeClustering

# from scipy .cluster.hierarchy import
```

```python
attributes,clusters = make_blobs()
```

```python
plt.scatter(attributes[:,0],attributes[:,1],c=clusters)
```

```python
k_means = KMeans(
    n_clusters=3,
    init = "random",
)
```

```python
k_means.fit(attributes)
```

```python
k_means.predict(attributes)
```

```python
clusters
```

```python
k_means.inertia_
```

```python
attributes,clusters = make_blobs(center_box = [-5,5], cluster_std=0.5)
```

```python
num_cluster = []
inertias=[]
for k in range(1,8):
    k_means= KMeans(n_clusters=k)
    k_means.fit(attributes)

    num_cluster.append(k)
    inertias.append(k_means.inertia_)
```

```python
plt.scatter(num_cluster,inertias)
plt.plot(num_cluster,inertias)
plt.xticks(range(1,max(num_cluster)+1))
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()
```
